```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mpl.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams["mathtext.default"] = 'regular'
mpl.rcParams['pdf.fonttype'] = 42
# This is for using seaborn
sns.set_theme(style="white")
sns.set_style('ticks')
mpl.rcParams['font.size'] = 12
```

```python
df = pd.read_excel("../xrd_data.xlsx")
df.head(2)
```

```python
# Set the index to the Samples column
df.set_index('Samples', inplace=True)

# Normalize the data (so that the proportions sum to 1)
df = df.div(df.sum(axis=1), axis=0)
```

```python
# Stacked Bar Chart
fig, ax = plt.subplots(figsize=(12,6))

# Specify the colors for each mineral
# colors = [
#     'blue',     # Phyllosilicates
#     'red',      # Quartz
#     'green',    # Feldspath-K
#     'purple',   # Plagioclase-Na
#     'orange',   # Calcite
#     'cyan',     # Dolomite
#     'magenta',  # Pyrite
#     'gray',     # Ankérite
#     'brown'     # Indosés
# ]
colors = [
    '#888888',
    '#1f77b4',  # Blue
    # '#ff7f0e',  # Orange
    '#2ca02c',  # Green
    '#d62728',  # Red
    '#87dee7',  # aqua
    '#e377c2',  # Pink
    'black',    # Black
    '#bcbd22'  # Olive
]

df.plot(kind='bar', stacked = True, ax=ax, color = colors)

plt.title('Mineralogical Composition')
plt.xlabel('Samples', color ='black', fontweight = 'bold', fontsize =12)
plt.ylabel('Proportions',color ='black', fontweight = 'bold', fontsize =12)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), edgecolor='black')
plt.tick_params(color='black',)
plt.savefig("../plots/xrd_data.pdf", dpi = 600)
```

```python
# Heatmap
fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(df, cmap='viridis', annot=False, fmt='.2f', ax=ax)
plt.title('Heatmap of Mineral Proportions')
plt.xlabel('Minerals')
plt.ylabel('Samples')
plt.show()
```
