```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Make global settings for the figures.
mpl.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams["mathtext.default"] = 'regular'
mpl.rcParams['pdf.fonttype'] = 42
# This is for using seaborn
sns.set_theme(style="white")
sns.set_style('ticks')
save_path ='../plots/'
```

```python
# import data into a pandas dataframe.
cha = pd.read_excel("../SSB_literature_Hg_data.xlsx", sheet_name="Chaohu")
cha.head(3)
```

```python
depth = cha.depth
D199 = cha.D199Hg
Hg = cha.Hg
Hg_TOC=cha.Hg_TOC
err199 = cha.D199_2se
```

```python
fig, axes = plt.subplots(figsize=(4, 7), nrows=1, ncols = 2, sharey=True)
ax0, ax1 = axes

ax_list = [ax0, ax0, ax1]
plt_list = [Hg, Hg_TOC, D199]
labels = ["Hg (ppb)", "Hg/TOC (ppb/wt.%)", "$\Delta^{199}Hg\;(‰)$"]
colors=["black", "white", "black"]
# remove the y-axis ticks and y-axis itself.
sns.despine(left=True, right=True, top=True, bottom=False)
plt.yticks([])

for i in range(len(ax_list)):
    sns.scatterplot(data=cha, x=plt_list[i], y=depth, ax=ax_list[i],
                    color=colors[i], legend=False, edgecolor='black',
                    size=20,
                    )
    ax_list[i].tick_params(labelsize=10)
    ax_list[i].set_xlabel(labels[i], fontsize=10, c="black", fontweight="bold")
ax1.errorbar(xerr=err199, y=depth, x=D199, ecolor="black", color="black",
                  linestyle='none', elinewidth=0.7, capthick=0.5, mec='black')
ax1.axvline(x=-.05, linestyle='--', color='black', linewidth=1)
ax1.axvline(x=.05, linestyle='--', color='black', linewidth=1)
ax1.set_xticks([-0.1, 0, 0.1,0.2])
# plt.savefig(f'{save_path}Chaohu.pdf')
```

```python
# Create figure
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(6, 10), sharex=True)
plot_axes = [ax0,ax1,ax2]
plot_data =["d202Hg", "D199Hg", "D200Hg"]
colors = ["fuchsia", "salmon", "#33cb9c", "#289cc9"]
labels = ["$\delta^{200}Hg\;(‰)$", "$\Delta^{199}Hg\;(‰)$", "$\Delta^{200}Hg\;(‰)$"]
for i in range(len(plot_axes)):
    sns.boxplot(x='Locality', y=plot_data[i], data=iso, ax=plot_axes[i],
                medianprops={"color": "black"}, flierprops={"marker": "o"},
                width=0.5, palette=colors)
    plot_axes[i].set_ylabel(labels[i], size=10, c="black", fontweight="bold")
    plot_axes[i].set_xlabel("Locality", size=10)
    plot_axes[i].tick_params(labelsize=10, color="black")

plt.savefig(f'{save_path}Hg-iso_boxplot.pdf')

# Create figure landscape mode
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, figsize=(11, 5), sharey=True)
plot_axes = [ax0, ax1, ax2]
plot_data = ["d202Hg", "D199Hg", "D200Hg"]

for i in range(len(plot_axes)):
    sns.boxplot(y='Locality', x=plot_data[i], data=iso, ax=plot_axes[i],
                medianprops={"color": "black"}, flierprops={"marker": "o"},
                width=0.5, palette=colors)
    plot_axes[i].set_xlabel(labels[i], size=10, c="black", fontweight="bold")
    plot_axes[i].set_ylabel("Locality", size=10)
    plot_axes[i].tick_params(labelsize=10, color="black")

plt.savefig(f'{save_path}Hg-iso_boxplot-lscp.pdf')
```
