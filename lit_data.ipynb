```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Make global settings for the figures.
mpl.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial']
plt.rcParams["mathtext.default"] = 'regular'
mpl.rcParams['pdf.fonttype'] = 42
# This is for using seaborn
sns.set_theme(style="white")
sns.set_style('ticks')

# declare paths to save to and extract data exce sheet files.
save_path ='../plots/'
data_path = "../SSB_literature_Hg_data.xlsx"
font_size = 12
```

```python
# import data into a pandas dataframe.
cha = pd.read_excel(data_path, sheet_name="Chaohu")
cha.head(3)
```

```python
depth = cha.depth
D199 = cha.D199Hg
Hg = cha.Hg
Hg_TOC=cha.Hg_TOC
err199 = cha.D199_2se
```

```python
fig, axes = plt.subplots(figsize=(3.4, 5.6), nrows=1, ncols=2, sharey=True)
ax0, ax1 = axes

ax_list = [ax0, ax0]
plt_list = [Hg, Hg_TOC]
labels = ["Hg (ppb)", "Hg/TOC (ppb/wt.%)"]
colors = ["black", "white"]
# remove the y-axis ticks and y-axis itself.
sns.despine(left=True, right=True, top=True, bottom=False)
plt.yticks([])

for i in range(len(ax_list)):
    sns.scatterplot(data=cha, x=plt_list[i], y=depth, ax=ax_list[i],
                    color=colors[i], legend=False, edgecolor='black',
                    size=20,
                    )
    ax_list[i].tick_params(labelsize=10)
    ax_list[i].set_xlabel(labels[i], fontsize=font_size, c="black", fontweight="bold")

ax0.set_xticks([0, 100,200])
plt.savefig(f'{save_path}WPND-corr.pdf')
```

```python
fig, axes = plt.subplots(figsize=(3.6, 5.6), nrows=1, ncols = 2, sharey=True)
ax0, ax1 = axes

ax_list = [ax0, ax0, ax1]
plt_list = [Hg, Hg_TOC, D199]
labels = ["Hg (ppb)", "Hg/TOC (ppb/wt.%)", "$\Delta^{199}Hg\;(‰)$"]
colors=["black", "white", "black"]
# remove the y-axis ticks and y-axis itself.
sns.despine(left=True, right=True, top=True, bottom=False)
plt.yticks([])

for i in range(len(ax_list)):
    sns.scatterplot(data=cha, x=plt_list[i], y=depth, ax=ax_list[i],
                    color=colors[i], legend=False, edgecolor='black',
                    size=20,
                    )
    ax_list[i].tick_params(labelsize=font_size)
    ax_list[i].set_xlabel(labels[i], fontsize=font_size, c="black", fontweight="bold")
ax1.errorbar(xerr=err199, y=depth, x=D199, ecolor="black", color="black",
                  linestyle='none', elinewidth=0.7, capthick=0.5, mec='black')
ax1.axvline(x=-.05, linestyle='--', color='black', linewidth=1)
ax1.axvline(x=.05, linestyle='--', color='black', linewidth=1)
ax0.set_xticks([0, 100, 200])
ax1.set_xticks([-0.1, 0, 0.1,0.2])
plt.savefig(f'{save_path}Chaohu.pdf')
```

```python
wall = pd.read_excel(data_path, sheet_name='Hammer_2019')
wall.head(2)
```

```python
Hg = wall.Hg
Hg_TOC = wall.Hg_TOC
d13C = wall.d13C
depth = wall.depth
```

```python
fig, axes = plt.subplots(figsize=(3.6, 5), nrows=1, ncols=2, sharey=True)
ax0, ax1 = axes
ax_list = [ax0, ax1]
plt_list = [Hg_TOC, d13C]
labels = ["Hg/TOC (ppb/wt.%)", "$\delta^{13}C\;(‰\;VPDB)$"]
colors = ["white", "#848482"]
# remove the y-axis ticks and y-axis itself.
sns.despine(left=True, right=True, top=True, bottom=False)
plt.yticks([])

def make_plot(data, f_size, f_name):
    for i in range(len(ax_list)):
        sns.scatterplot(data=data, x=plt_list[i], y=depth, ax=ax_list[i],
                        color=colors[i], legend=False, edgecolor='black',
                        size=20,
                        )
        ax_list[i].tick_params(labelsize=f_size)
        ax_list[i].set_xlabel(labels[i], fontsize=f_size, c="black", fontweight="bold")
    plt.savefig(f'{save_path}{f_name}.pdf')

make_plot(wall, font_size, 'wall-corr');
```
